```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset=pd.read_csv("kidney_disease.csv")
df2=pd.DataFrame(dataset)
df2
```

```python
df2.info()
```

```python
df2.isnull().sum()
```

```python
df2.drop(['id'], axis=1, inplace=True)
```

```python
df2["pcv"].unique()
```

```python
df2["wc"].unique()
```

```python
df2["rc"].unique()
```

```python
df2["pcv"] = pd.to_numeric(df2["pcv"], errors="coerce")
df2["wc"] = pd.to_numeric(df2["wc"], errors="coerce")
df2["rc"] = pd.to_numeric(df2["rc"], errors="coerce")
```

```python
for col in df2.columns:
    if df2[col].dtype == "object":
        df2[col].fillna(df2[col].mode()[0], inplace=True)
    else:
        df2[col].fillna(df2[col].median(), inplace=True)
df2.isnull().sum()
```

```python
df2.info()
```

```python
df2.duplicated().sum()
```

```python
numeric_cols = df2.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df2.select_dtypes(include=['object', 'category']).columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in numeric_cols:
    plt.figure(figsize=(14, 5))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df2[col], kde=True,palette="viridis")
    plt.title(f"Histogram of {col}")

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df2[col],palette="pastel")
    plt.title(f"Boxplot of {col}")

    plt.show()
```

```python
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=df2[col],palette="Set2")
    plt.title(f"Count Plot of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2["rbc"]=le.fit_transform(df2["rbc"])
df2["pc"]=le.fit_transform(df2["pc"])
df2["pcc"]=le.fit_transform(df2["pcc"])
df2["ba"]=le.fit_transform(df2["ba"])
df2["htn"]=le.fit_transform(df2["htn"])
df2["dm"]=le.fit_transform(df2["dm"])
df2["cad"]=le.fit_transform(df2["cad"])
df2["pe"]=le.fit_transform(df2["pe"])
df2["ane"]=le.fit_transform(df2["ane"])
df2["classification"]=le.fit_transform(df2["classification"])
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
df2["appet"]= oe.fit_transform(df2[["appet"]])
```

```python
x=df2.drop("classification",axis=1)
y=df2["classification"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
x_smote,y_smote=smote.fit_resample(x,y)
y_smote.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
scaler2=StandardScaler()
x_train=scaler2.fit_transform(x_train)
x_test=scaler2.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
knn2=KNeighborsClassifier(n_neighbors=20)
knn2.fit(x_train,y_train)
```

```python
knn2.score(x_test,y_test)
```

```python
y_hat=knn2.predict(x_test)
accuracy_score(y_test,y_hat)
print(classification_report(y_test,y_hat))
cm =confusion_matrix(y_test,y_hat)
plt.figure(figsize=(6,4))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
lr2=LogisticRegression()
lr2.fit(x_train,y_train)
```

```python
lr2.score(x_test,y_test)
```

```python
y_hat=lr2.predict(x_test)
accuracy_score(y_test,y_hat)
print(classification_report(y_test,y_hat))
cm =confusion_matrix(y_test,y_hat)
plt.figure(figsize=(6,4))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc2=DecisionTreeClassifier()
dtc2.fit(x_train,y_train)
```

```python
dtc2.score(x_test,y_test)
```

```python
y_hat=dtc2.predict(x_test)
accuracy_score(y_test,y_hat)
print(classification_report(y_test,y_hat))
cm =confusion_matrix(y_test,y_hat)
plt.figure(figsize=(6,4))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc2=RandomForestClassifier()
rfc2.fit(x_train,y_train)
```

```python
rfc2.score(x_test,y_test)
```

```python
y_hat=rfc2.predict(x_test)
accuracy_score(y_test,y_hat)
print(classification_report(y_test,y_hat))
cm =confusion_matrix(y_test,y_hat)
plt.figure(figsize=(6,4))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.metrics import accuracy_score

# Predictions for all models
models = {
    "Logistic Regression": lr2,
    "K-Nearest Neighbors": knn2,
    "Decision Tree": dtc2,
    "Random Forest": rfc2
}

# Create empty list to store results
results = []

# Evaluate each model
for name, model in models.items():
    y_hat = model.predict(x_test)
    score = accuracy_score(y_test, y_hat)
    results.append({"Model": name, "Score": round(score, 4)})

# Create summary DataFrame
models_df = pd.DataFrame(results)

# Display neatly
print("Model Comparison:")
display(models_df.sort_values(by="Score", ascending=False))
```
