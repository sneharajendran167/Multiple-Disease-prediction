```python
df["Dataset"]=df["Dataset"].map({1:0,2:1})
```

```python
x=df.drop('Dataset',axis=1)
y=df['Dataset']
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
x_smote,y_smote=smote.fit_resample(x,y)
y_smote.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_hat=lr.predict(x_test)
print(classification_report(y_test,y_hat))
cm = confusion_matrix(y_test,y_hat)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=24)
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_hat=knn.predict(x_test)
print(classification_report(y_test,y_hat))
cm = confusion_matrix(y_test,y_hat)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(random_state=42)
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_hat=dt.predict(x_test)
print(classification_report(y_test,y_hat))
cm = confusion_matrix(y_test,y_hat)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.ensemble import BaggingClassifier
bg=BaggingClassifier(estimator=dt, n_estimators=50, random_state=42)
bg.fit(x_train,y_train)
```

```python
bg.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_hat=bg.predict(x_test)
print(classification_report(y_test,y_hat))
cm = confusion_matrix(y_test,y_hat)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train,y_train)
```

```python
rf.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_hat=rf.predict(x_test)
print(classification_report(y_test,y_hat))
cm = confusion_matrix(y_test,y_hat)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.metrics import accuracy_score

# Predictions for all models
models = {
    "Logistic Regression": lr,
    "K-Nearest Neighbors": knn,
    "Decision Tree": dt,
    "Bagging Classifier": bg,
    "Random Forest": rf
}

# Create empty list to store results
results = []

# Evaluate each model
for name, model in models.items():
    y_hat = model.predict(x_test)
    score = accuracy_score(y_test, y_hat)
    results.append({"Model": name, "Score": round(score, 4)})

# Create summary DataFrame
models_df = pd.DataFrame(results)

# Display neatly
print("Model Comparison:")
display(models_df.sort_values(by="Score", ascending=False))
```
